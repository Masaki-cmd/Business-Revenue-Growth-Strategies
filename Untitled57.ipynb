## 課題定義と仮説構築

### Subtask:
クリーニングされたデータから、収益（売上、利益、コスト）に影響を与えている可能性のある課題や問題点を特定します。例えば、「特定の製品の売上が低い」「特定地域の顧客離反率が高い」など。これらの課題に対して、収益向上に繋がる具体的な仮説（例: 「製品Xの価格を調整すれば売上が向上する」）を日本語で構築します。


**Reasoning**:
First, I'll extract month and year from `Order Date` to create a new time-based column. Then, I'll aggregate the data by this new time column, `Category`, and `Sub-Category` to get the sum of `Amount`, `Profit`, and `Quantity`. This aligns with the first instruction of identifying trends and anomalies by time and product categories.



````python
df_merged['Order_Month_Year'] = df_merged['Order Date'].dt.to_period('M')

df_aggregated_time_category = df_merged.groupby(['Order_Month_Year', 'Category', 'Sub-Category']).agg(
    Amount=('Amount', 'sum'),
    Profit=('Profit', 'sum'),
    Quantity=('Quantity', 'sum')
).reset_index()

print("時間、カテゴリ、サブカテゴリ別の主要メトリクス集計の最初の5行:")
print(df_aggregated_time_category.head())

print("\n時間、カテゴリ、サブカテゴリ別の主要メトリクス集計の情報:")
print(df_aggregated_time_category.info())
````

**Reasoning**:
To identify potential issues and formulate hypotheses, I will analyze the `df_aggregated_time_category` DataFrame, focusing on identifying sub-categories that consistently show negative profit. This will help pinpoint areas for improvement or further investigation.



````python
negative_profit_subcategories = df_aggregated_time_category[df_aggregated_time_category['Profit'] < 0]

# サブカテゴリごとの合計利益を計算し、マイナス利益のものを抽出
negative_profit_summary = negative_profit_subcategories.groupby(['Category', 'Sub-Category'])['Profit'].sum().reset_index()
negative_profit_summary = negative_profit_summary.sort_values(by='Profit', ascending=True)

print("時間、カテゴリ、サブカテゴリ別の負の利益を持つサブカテゴリ:\n")
print(negative_profit_subcategories.head())

print("\n合計利益が負のサブカテゴリの要約 (損失が大きい順):\n")
print(negative_profit_summary.to_markdown(index=False, numalign="left", stralign="left"))
````

## 詳細データ分析と可視化

### Subtask:
構築した仮説を検証するために、データ分析を行います。適切な統計手法や機械学習モデル（必要に応じて）を使用し、課題の原因を深く掘り下げます。分析結果はグラフを用いて可視化し、「なぜこのグラフを作成したのか」「この結果からなぜその提案が導かれるのか」という論理的な説明を日本語で記述します。


**Reasoning**:
First, I'll filter the `df_merged` DataFrame to include only the sub-categories that had negative total profit, as identified in `negative_profit_summary`. Then, I'll group this filtered data by `Category` and `Sub-Category` to calculate the total `Amount`, `Profit`, and `Quantity` for each. Finally, I will calculate 'Profit Margin' (Profit / Amount) and 'Profit per Unit' (Profit / Quantity) for each of these sub-categories to understand the source of the losses, addressing the first instruction of the subtask.



````python
negative_subcategories_list = negative_profit_summary['Sub-Category'].unique()

df_negative_subcategories = df_merged[df_merged['Sub-Category'].isin(negative_subcategories_list)].copy()

negative_profit_metrics = df_negative_subcategories.groupby(['Category', 'Sub-Category']).agg(
    Total_Amount=('Amount', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Total_Quantity=('Quantity', 'sum')
).reset_index()

negative_profit_metrics['Profit_Margin'] = (negative_profit_metrics['Total_Profit'] / negative_profit_metrics['Total_Amount']) * 100
negative_profit_metrics['Profit_Per_Unit'] = negative_profit_metrics['Total_Profit'] / negative_profit_metrics['Total_Quantity']

# 損失額でソートし、必要な列を表示
negative_profit_metrics = negative_profit_metrics.sort_values(by='Total_Profit', ascending=True)

print("合計利益が負のサブカテゴリにおける詳細指標:")
print(negative_profit_metrics.to_markdown(index=False, numalign="left", stralign="left"))
````

**Reasoning**:
To better understand the magnitude and nature of the losses, I will visualize the 'Profit Margin' and 'Profit per Unit' for the identified negative-profit sub-categories. This will help to validate the first hypothesis regarding pricing strategy and cost optimization by showing which products have the worst margins and per-unit profitability.



````python
import matplotlib.pyplot as plt
import seaborn as sns

# 損失が最も大きい上位N個のサブカテゴリに絞ることも可能ですが、今回は全て可視化します
# negative_profit_metrics_plot = negative_profit_metrics.head(10) # 例: 上位10件
negative_profit_metrics_plot = negative_profit_metrics[negative_profit_metrics['Total_Profit'] < 0]

# FigureとAxesの準備
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# Profit Marginの可視化
sns.barplot(x='Profit_Margin', y='Sub-Category', data=negative_profit_metrics_plot.sort_values(by='Profit_Margin'), ax=axes[0], palette='viridis')
axes[0].set_title('損失サブカテゴリにおける利益率 (%): 低い順')
axes[0].set_xlabel('利益率 (%)')
axes[0].set_ylabel('サブカテゴリ')
axes[0].grid(axis='x', linestyle='--', alpha=0.7)

# Profit Per Unitの可視化
sns.barplot(x='Profit_Per_Unit', y='Sub-Category', data=negative_profit_metrics_plot.sort_values(by='Profit_Per_Unit'), ax=axes[1], palette='plasma')
axes[1].set_title('損失サブカテゴリにおけるユニットあたり利益: 低い順')
axes[1].set_xlabel('ユニットあたり利益')
axes[1].set_ylabel('サブカテゴリ')
axes[1].grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
````

# README: 売上・利益向上分析プロジェクト

## 概要

このプロジェクトは、提供された販売データ（`List of Orders.csv`, `Order Details.csv`, `Sales target.csv`）を用いて、企業の売上と利益を向上させるための分析を行いました。主な目的は、収益最大化またはコスト最適化をKGIとし、データから具体的なアクションプランと投資対効果（ROI）を導き出すことです。

## データソース

-   **`List of Orders.csv`**: 注文ID、注文日、顧客名、州、都市など、注文に関する基本情報。
-   **`Order Details.csv`**: 注文ID、金額、利益、数量、カテゴリ、サブカテゴリなど、各注文の詳細情報。
-   **`Sales target.csv`**: 月ごとのカテゴリ別売上目標。

## 分析目的

1.  **問題特定と仮説構築**: 収益に影響を与える課題を特定し、それに対する仮説を構築する。
2.  **詳細データ分析と可視化**: 構築した仮説を検証し、課題の根本原因を深く掘り下げる。
3.  **アクションプランとROI評価**: 分析結果に基づき、収益向上またはコスト最適化のための具体的なアクションプランと投資対効果（ROI）を提案する。

## 分析プロセスと主要な発見

### 1. データ準備とクリーニング

-   `archive.zip`を解凍し、3つのCSVファイルを読み込みました。
-   `List of Orders.csv`から60件の欠損値（全列）と59件の重複行を削除し、`Order Date`を日付型に変換しました。

    2.  **在庫管理とプロモーションの見直し**: 損失を出している製品群の在庫最適化とダイナミックなプロモーションにより、損失を削減する。
    3.  **サプライヤー交渉とコスト削減**: 仕入れコストまたは製造コストの削減により、利益率を改善する。
